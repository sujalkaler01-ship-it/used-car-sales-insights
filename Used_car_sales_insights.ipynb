## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Create a copy of the original dataset
cars = df.copy()

# Remove duplicate rows
cars = cars.drop_duplicates()

# Fill missing categorical values with "Unknown"
categorical_cols = ["zipcode", "Make", "Model", "Trim", "Engine", "BodyType", "DriveType"]
for col in categorical_cols:
    if col in cars.columns:
        cars[col] = cars[col].fillna("Unknown")

# Clean BodyType because it has many messy/special-character values
cars["BodyType"] = cars["BodyType"].astype(str).str.replace(r"[^A-Za-z0-9\s]", "", regex=True)
cars["BodyType"] = cars["BodyType"].str.strip()
cars["BodyType"] = cars["BodyType"].replace("", "Unknown")

# Create a new feature: Car_Age
# Car_Age means how old the car is compared to the selling/listing year
cars["Car_Age"] = cars["yearsold"] - cars["Year"]

# Remove unrealistic rows
cars = cars[cars["pricesold"] > 0]
cars = cars[cars["Mileage"] >= 0]
cars = cars[cars["Car_Age"] >= 0]

# Show cleaned dataset
cars.head()
````

### What all manipulations have you done and insights you found?

I removed duplicate rows, handled missing values, cleaned the BodyType column, and created a new column called Car_Age. I also removed unrealistic rows where selling price, mileage, or car age values were not valid. After cleaning, the dataset became easier to analyze and visualize. The main insight is that price, mileage, age, brand, body type, and drive type are important factors for used car analysis.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
# Correlation between numerical columns

numeric_cols = ["pricesold", "yearsold", "Mileage", "Year", "NumCylinders", "Car_Age"]
corr = cars[numeric_cols].corr()

plt.figure(figsize=(8, 6))
plt.imshow(corr, aspect="auto")
plt.colorbar(label="Correlation")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")

# Show correlation values inside the heatmap
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(j, i, round(corr.iloc[i, j], 2), ha="center", va="center")

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is used because it quickly shows relationships between numerical variables.

##### 2. What is/are the insight(s) found from the chart?

The heatmap helps identify how selling price is related to mileage, year, car age, and cylinders.
